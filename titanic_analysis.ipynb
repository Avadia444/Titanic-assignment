Survival probability of a females = the number of females who survived /Total number of males and females
Survival probability of a males = the number of males who survived /Total number of males and females

```python
#choosing only the survived ones
df1=df[df['survived']==1].groupby(by='sex').count()
df1
```

```python
s_f=df1['survived'].loc['female']
s_m=df1['survived'].loc['male']
t_num=df1['survived'].sum()
t_num
```

```python
df[df['survived']==1].count()['sex']
```

```python
p_f=s_f/t_num
p_f
```

```python
p_m=s_m/t_num
p_m
```

*** 2. Calculate the same proportion, but by class and sex. (10 pts)*** 

Survival probability by sex and class = the number of people in that sex and class who survived/Total number of people in all categories of sex and class

```python
df[df.survived==1].groupby(by=['pclass','sex']).survived.count().loc[1]/df[df.survived==1].groupby(['pclass','sex']).survived.count().loc[1].sum()
```

```python
df[df.survived==1].groupby(by=['pclass','sex']).survived.count().loc[2]/df[df.survived==1].groupby(['pclass','sex']).survived.count().loc[2].sum()
```
